# [dbo].[spUsuarioInsert]

```sql
/****** Object:  StoredProcedure [dbo].[spUsuarioInsert]    Script Date: 28/08/2023 7:05:16 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE    PROCEDURE  [dbo].[spUsuarioInsert]
(
	@Email varchar(250),
	@Password varchar(30)
)
AS
BEGIN 

	SET NOCOUNT ON;
	BEGIN TRY
		DECLARE @PasswordHash varbinary(32);
		select @PasswordHash= HASHBYTES('SHA2_256',@Password);

		INSERT INTO dbo.TblUsuario
		(FechaAlta, Email,Password)
		values(GETDATE(),@email,@PasswordHash);
	END TRY

	BEGIN CATCH
		execute dbo.sp_ControlErrores;
	END CATCH
END 
GO
```

# [dbo].[spUsuarioSetPassword]

```sql
/****** Object:  StoredProcedure [dbo].[spUsuarioSetPassword]    Script Date: 28/08/2023 7:05:16 p. m. ******/
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO

CREATE   PROCEDURE [dbo].[spUsuarioSetPassword]
(
	@Email varchar(250),
	@OldPassword varchar(30),
	@NewPassword varchar(30)
)
AS
BEGIN
	SET NOCOUNT ON;
	BEGIN TRY
	
		DECLARE @OldHashSegunUsuario varbinary(32);
		SELECT  @OldHashSegunUsuario= HASHBYTES('SHA2_256',@OldPassword);

		DECLARE  @Coincide int =0;
		SELECT  @Coincide=COUNT(*) FROM dbo.TblUsuario where Password=@OldHashSegunUsuario;
		IF @Coincide=0
		BEGIN 
			SELECT 0; RETURN;
		END

		--si ha coincidido, se establece un nuevo password
		DECLARE @NuevoHash varbinary (32);
		SELECT @NuevoHash= HASHBYTES('SHA2_256', @NewPassword);
		UPDATE dbo.TblUsuario set Password=@NuevoHash WHERE  Email=@Email;
		SELECT 1; RETURN;
	END TRY
	BEGIN CATCH
		execute dbo.sp_ControlErrores;
	END CATCH
END
GO
```
